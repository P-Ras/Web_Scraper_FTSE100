```python
import requests
from bs4 import BeautifulSoup
```

```python
#Use the LSE URLs which shows the list of company in FTSE 100 with heatmap format
URL = 'https://www.londonstockexchange.com/indices/ftse-100/constituents/heatmap?lang=en'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
results = soup.find_all(class_='map-instrument flex-container positivechange')
```

```python
results
```

```python
for symbol_row in results:
    symbol = symbol_row.find('div', class_='flex-container map-data')
    print(symbol.text.strip())
```
